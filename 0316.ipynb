```python
#사칙연산의 결과 값을 반환하는 함수
덧셈: add(a,b)
뺄셈: diff(a,b)
나눗셈: div(a,b)
곱셈: mul(a,b)
```

```python
def add (a,b):
    return a + b
    
def diff(a,b):
    return a - b

def div(a,b):
    return a / b

def mul(a,b):
    return a * b
    
```
